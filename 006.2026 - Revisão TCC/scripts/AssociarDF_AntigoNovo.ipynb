```python
import requests, time, json, os

def consultar_porte_cnpj(cnpjs, repo_path, sleep=0.3, salvar_cada=50):
    """
    Consulta o porte tributário (Receita Federal via BrasilAPI) para uma lista de CNPJs.
    Retorna dict {cnpj_14dig: porte}. Usa cache em disco para não reconsultar.

    porte possíveis: 'MICRO EMPRESA', 'EMPRESA DE PEQUENO PORTE', 'DEMAIS', None
    """
    cache_path = os.path.join(repo_path, 'outputs', 'cache_porte_cnpj.json')
    if os.path.exists(cache_path):
        with open(cache_path, 'r', encoding='utf-8') as f:
            cache = json.load(f)
    else:
        cache = {}

    # normaliza p/ 14 dígitos e remove inválidos/duplicados
    cnpjs = (
        pd.Series(cnpjs).dropna().astype(str)
        .str.replace(r'\D', '', regex=True)
        .loc[lambda s: s.str.len() == 14]
        .unique()
    )

    session = requests.Session()
    for i, cnpj in enumerate(cnpjs, 1):
        if cnpj in cache:          # já consultado antes
            continue
        url = f'https://brasilapi.com.br/api/cnpj/v1/{cnpj}'
        for tentativa in range(3):  # retry com backoff
            try:
                r = session.get(url, timeout=20)
                if r.status_code == 200:
                    cache[cnpj] = r.json().get('porte')
                    break
                elif r.status_code == 429:           # rate limit
                    time.sleep(2 ** tentativa)
                else:                                 # 404 etc.
                    cache[cnpj] = None
                    break
            except requests.RequestException:
                time.sleep(2 ** tentativa)
        else:
            cache[cnpj] = None

        if i % salvar_cada == 0:
            print(f'{i}/{len(cnpjs)} consultados')
            with open(cache_path, 'w', encoding='utf-8') as f:
                json.dump(cache, f, ensure_ascii=False, indent=2)
        time.sleep(sleep)           # respeita rate limit da BrasilAPI

    with open(cache_path, 'w', encoding='utf-8') as f:
        json.dump(cache, f, ensure_ascii=False, indent=2)
    return cache

mapa_porte = consultar_porte_cnpj(verif01_newer_classif['CNPJ'], repo_path)

verif01_newer_classif['porte_industria'] = (
    verif01_newer_classif['CNPJ']
    .astype(str).str.replace(r'\D', '', regex=True).str.rjust(14, '0')
    .map(mapa_porte)
)

verif01_newer_classif['porte_industria'].value_counts(dropna=False)
```
